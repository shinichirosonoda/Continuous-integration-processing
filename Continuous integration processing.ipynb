```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,101)

m = 3
n = 2

A = 1
B = 0.8

ph = np.pi/60

y1 = np.sin(m * x)
y2 = np.sin(n *x)

y3 = A * y1 + B * y2
y4 = np.sin(m *(x + ph))

plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y1)
plt.plot(x,y2)
```

$
0 \leq x \leq 2\pi \\ 
y = sin(mx) \\
y = sin(nx) \\
m, n \in\mathbb{N}
$

```python
x = np.linspace(-np.pi, np.pi,101)
y1 = np.sin(m * (x - np.pi))
y2 = np.sin(n * (x - np.pi))

y3 = A * y1 + B * y2
y4 = np.sin(m *(x + ph))

plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y1)
plt.plot(x,y3)
```

$
-\pi \leq x \leq \pi \\
\begin{eqnarray}
f(x) &=& sin(m(x - \pi)) \\
  &=& (-1)^{m}sin(mx) \\
g(x) &=& sin(n(x - \pi)) \\
  &=& (-1)^{n}sin(nx) \\
\end{eqnarray}\\
m, n \in\mathbb{N} においてf(x), g(x)は奇関数となる。
$

$
A, B \in\mathbb{R} \\
h(x) = Af(x) + Bg(x)
とするとf(x), g(x)は奇関数よりh(x)も奇関数となる。
$

$
m = 2M + 1 \quad M \in\mathbb{N} \\
\alpha < 0においてf(\alpha)=0を満たす\alpha_{i-} = -\pi + \frac{2\pi i}{2M + 1} \quad (i = 0,1,2,...,M) \\
\alpha > 0においてf(\alpha)=0を満たす\alpha_{i+} =  \pi - \frac{2\pi i}{2M + 1} \quad (i = 0,1,2,...,M) \\ 
となる。\\
\beta < 0において \\
h(\beta)=0を満たす\beta_{i-} \quad (i = 0,1,2,...,M') の中で \\
\alpha_{i-}に最も距離が近い\beta_{i-}を\beta'_{i-} \quad (i = 0,1,2,...,M) とする。\\
\beta > 0において \\
h(\beta)=0を満たす\beta_{i+} \quad (i = 0,1,2,...,M') の中で \\
\alpha_{i+}に最も距離が近い\beta_{i+}を\beta'_{i+} \quad (i = 0,1,2,...,M) とする。\\
$

$
f(x)とh(x)の積算位相差\psiを \\
\psi = \sum_{n=0}^{M}((\beta'_{i-} -　\alpha_{i-}) + (\beta'_{i+} -　\alpha_{i+})) \\
f(x), h(x)は奇関数より\\
f(x) = -f(-x) = 0 より \alpha_{i+} = -\alpha_{i-} \\
h(x) = -h(-x) = 0 より \beta'_{i+} = -\beta'_{i-} \\
よって
\psi=0となる。
$

```python
m = 2
n = 3
y1 = np.sin(m * (x - np.pi))
y2 = np.sin(n * (x - np.pi))

y3 = A * y1 + B * y2
y4 = np.sin(m *(x + ph))

plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y1)
plt.plot(x,y3)
```

$
m = 2M \quad M \in\mathbb{N} \\
\alpha < 0においてf(\alpha)=0を満たす\alpha_{i-} = -\pi + \frac{2\pi i}{2M} \quad (i = 0,1,2,...,M-1) \\
\alpha > 0においてf(\alpha)=0を満たす\alpha_{i+} =  \pi - \frac{2\pi i}{2M} \quad (i = 0,1,2,...,M-1) \\
f(\alpha_{M+}) = f(\alpha_{M-})=0を満たす。\alpha_{M+}=\alpha_{M-}=0 \\
となる。\\
\beta < 0において \\
h(\beta)=0を満たす\beta_{i-} \quad (i = 0,1,2,...,M'-1) の中で \\
\alpha_i-に最も距離が近い\beta_{i-}を\beta'_{i-} \quad (i = 0,1,2,...,M-1) とする。\\
\beta > 0において \\
h(\beta)=0を満たす\beta_{i+} \quad (i = 0,1,2,...,M'-1) の中で \\
\alpha_{i+}に最も距離が近い\beta_{i+}を\beta'_{i+} \quad (i = 0,1,2,...,M-1) とする。\\
★h(\beta'_{M+}) = h(\beta'_{M-})=0を満たす、\beta'_{M+}=\beta'_{M-}=0 の場合、\\
すなわち\alpha_{M+}及び\alpha_{M-}に最も近い\beta'_{M+},\beta'_{M-}が0の時を考える。
$

$
f(x)とh(x)の積算位相差\psiを \\
\psi = (\beta'_{M+} -\alpha_{M+}) + \sum_{n=0}^{M-1}((\beta'_{i-} -　\alpha_{i-}) + (\beta'_{i+} -　\alpha_{i+})) \\
f(x), h(x)は奇関数より\\
f(x) = -f(-x) = 0 より \alpha_{i+} = -\alpha_{i-} \\
h(x) = -h(-x) = 0 より \beta'_{i+} = -\beta'_{i-} \\
よって
\psi =0となる。
$

```python
x = np.linspace(-1.5*np.pi,np.pi,101)
m = 3
n = 2
y1 = np.sin(m * (x - np.pi))
y2 = np.sin(n * (x - np.pi))

y3 = A * y1 + B * y2
y4 = np.sin(m *((x + ph) - np.pi))

plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y4)
plt.plot(x,y3)
```

$
実際の測定はh(x)とf(x+\phi）の位相差を比較する。\\
この時、h(x)の位相揺らぎが\phiより小さければ、\\
h(x)とf(x)の積算位相差測定におけるそれぞれの測定点に位相差\phiのシフトが加算される。\\
m = 2M + 1 \quad M \in\mathbb{N} \\
\begin{eqnarray}
\psi &=& \sum_{n=0}^{M}((\phi'_{i-} -　\alpha_{i-}) + (\phi'_{i+} -　\alpha_{i+})) \\
   &=& (2M+2)\phi \\
\end{eqnarray}\\
測定個数2M+2で割った平均の位相差は\phiになる。\\
$

\end{eqnarray}\\
